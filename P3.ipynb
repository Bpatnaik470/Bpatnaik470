## Hypothesis Testing

### Subtask:
Formulate and test specific hypotheses related to factors influencing customer acquisition/engagement. This includes performing a t-test to compare 'visits' based on 'gender', conducting ANOVA to see if 'visits' differ significantly across different 'regions', and calculating correlations between numerical variables like 'income', 'age', 'school' and 'visits'.


**Reasoning**:
To begin the hypothesis testing, I will import the necessary functions: `ttest_ind` and `f_oneway` from `scipy.stats` for t-tests and ANOVA, and `pearsonr` from `scipy.stats` for correlation calculations.



```python
from scipy.stats import ttest_ind, f_oneway, pearsonr
print("Imported ttest_ind, f_oneway, and pearsonr from scipy.stats.")
```

**Reasoning**:
Following the instructions, I will first perform an independent samples t-test to compare 'visits' between different genders. Then, I will conduct a one-way ANOVA test to assess if 'visits' differ significantly across various regions. Finally, I will calculate Pearson correlation coefficients and p-values for 'visits' against 'income', 'age', and 'school'.



```python
print("\n--- Hypothesis Testing ---")

# 1. T-test for Gender
visits_male = df[df['gender'] == 'male']['visits']
visits_female = df[df['gender'] == 'female']['visits']
t_stat_gender, p_val_gender = ttest_ind(visits_male, visits_female, equal_var=False) # Assuming unequal variances
print(f"\nGender (T-test):\nT-statistic: {t_stat_gender:.3f}, P-value: {p_val_gender:.3f}")

# 2. ANOVA for Region
# Get unique regions and create a list of 'visits' for each region
region_groups = [df['visits'][df['region'] == region] for region in df['region'].unique()]
f_stat_region, p_val_region = f_oneway(*region_groups)
print(f"\nRegion (ANOVA):\nF-statistic: {f_stat_region:.3f}, P-value: {p_val_region:.3f}")

# 3. Correlation for Numerical Variables
print("\nCorrelation with Visits:")

# Visits vs Income
corr_income, p_val_income = pearsonr(df['visits'], df['income'])
print(f"  Income: Correlation = {corr_income:.3f}, P-value = {p_val_income:.3f}")

# Visits vs Age
corr_age, p_val_age = pearsonr(df['visits'], df['age'])
print(f"  Age: Correlation = {corr_age:.3f}, P-value = {p_val_age:.3f}")

# Visits vs School
corr_school, p_val_school = pearsonr(df['visits'], df['school'])
print(f"  School: Correlation = {corr_school:.3f}, P-value = {p_val_school:.3f}")
```

```python
print("\n--- Visualizing Hypothesis Test Results ---")

# Plot for Gender (T-test)
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='visits', data=df)
plt.title('Distribution of Visits by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Visits')
plt.text(0.05, 0.95, f'T-test P-value: {p_val_gender:.3f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.show()

# Plot for Region (ANOVA)
plt.figure(figsize=(10, 6))
sns.boxplot(x='region', y='visits', data=df)
plt.title('Distribution of Visits by Region')
plt.xlabel('Region')
plt.ylabel('Number of Visits')
plt.text(0.05, 0.95, f'ANOVA P-value: {p_val_region:.3f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.show()

print("Visualizations for gender and region hypothesis tests displayed.")
```
