```python
import pandas as pd

# Load the movies dataset
movies_df = pd.read_csv('/content/movies.csv')

# Load the ratings dataset
ratings_df = pd.read_csv('/content/ratings.csv')

print("Movies DataFrame head:")
display(movies_df.head())

print("\nRatings DataFrame head:")
display(ratings_df.head())
```

```python
# Merge both dataframes on 'movieId'
merged_df = pd.merge(ratings_df, movies_df, on='movieId', how='left')

print("Merged DataFrame head:")
display(merged_df.head())
```

```python
# Create User-Item Matrix using pivot_table
user_item_matrix = merged_df.pivot_table(index='userId', columns='title', values='rating')

print("User-Item Matrix head:")
display(user_item_matrix.head())
```

```python
# Fill row-wise NaNs with the corresponding user's mean ratings
user_item_matrix_filled = user_item_matrix.apply(lambda x: x.fillna(x.mean()), axis=1)

print("User-Item Matrix after filling NaNs (first 5 rows):")
display(user_item_matrix_filled.head())
```

```python
# Calculate Pearson correlation between users
# We transpose the matrix to calculate correlation between rows (users)
user_correlation_matrix = user_item_matrix_filled.T.corr(method='pearson')

print("User Correlation Matrix head:")
display(user_correlation_matrix.head())
```

```python
# Choose the correlation of all users with only User 1
user1_correlations = user_correlation_matrix[1]

# Sort the User 1 correlation in descending order
sorted_user1_correlations = user1_correlations.sort_values(ascending=False)

# Drop the NaN values generated in the correlation matrix
sorted_user1_correlations = sorted_user1_correlations.dropna()

print("Sorted User 1 Correlations (after dropping NaNs):")
display(sorted_user1_correlations.head())
```

```python
# Choose the top 50 users that are highly correlated to User 1
# Exclude User 1 itself from the list
top_50_users = sorted_user1_correlations[1:].head(50)

print("Top 50 Users correlated with User 1:")
display(top_50_users)
```

```python
# Predict the rating that User 1 might give for the movie with movieid 32

# Get the movie title for movieid 32
movie_title_32 = movies_df[movies_df['movieId'] == 32]['title'].iloc[0]
print(f"Movie ID 32 corresponds to movie: {movie_title_32}")

# Get ratings of the top 50 correlated users for this specific movie
top_50_users_ratings = user_item_matrix.loc[top_50_users.index, movie_title_32]

# Drop NaN values from these ratings (users who haven't rated this movie)
valid_ratings = top_50_users_ratings.dropna()

# Get the corresponding correlations for these valid ratings
valid_correlations = top_50_users.loc[valid_ratings.index]

# Calculate weighted average for prediction
if not valid_ratings.empty:
    predicted_rating = (valid_ratings * valid_correlations).sum() / valid_correlations.sum()
    print(f"\nPredicted rating for User 1 for movie '{movie_title_32}': {predicted_rating:.2f}")
else:
    print(f"\nCannot predict rating for movie '{movie_title_32}' for User 1 as none of the top 50 correlated users have rated this movie.")
```

```python
from surprise import Dataset, Reader, KNNBasic, SVD, NMF
from surprise.model_selection import cross_validate

# A Reader is needed to parse the file or DataFrame. Here, we'll use the already loaded merged_df.
reader = Reader(rating_scale=(0.5, 5.0)) # Assuming ratings are between 0.5 and 5.0

# Load the merged_df into a Surprise Dataset
data = Dataset.load_from_df(merged_df[['userId', 'movieId', 'rating']], reader)

print("Dataset loaded for Surprise library.")
```

### KNNBasic Model

```python
# Configure similarity options for KNNBasic
sim_options_msd = {'name': 'msd', 'user_based': True} # Using Mean Squared Difference similarity

# Initialize KNNBasic with 20 neighbors and MSD similarity
knn_basic = KNNBasic(k=20, sim_options=sim_options_msd, verbose=False)

# Cross-validate KNNBasic
print("Cross-validating KNNBasic...")
knn_basic_results = cross_validate(knn_basic, data, measures=['RMSE'], cv=5, verbose=False)

print(f"\nKNNBasic - Mean RMSE: {knn_basic_results['test_rmse'].mean():.3f}")
print(f"KNNBasic - All RMSE scores: {knn_basic_results['test_rmse']}")
```

### SVD Model

```python
# Initialize SVD model
svd = SVD(verbose=False)

# Cross-validate SVD
print("\nCross-validating SVD...")
svd_results = cross_validate(svd, data, measures=['RMSE'], cv=5, verbose=False)

print(f"SVD - Mean RMSE: {svd_results['test_rmse'].mean():.3f}")
print(f"SVD - All RMSE scores: {svd_results['test_rmse']}")
```

### NMF Model

```python
# Initialize NMF model
nmf = NMF(verbose=False)

# Cross-validate NMF
print("\nCross-validating NMF...")
nmf_results = cross_validate(nmf, data, measures=['RMSE'], cv=5, verbose=False)

print(f"NMF - Mean RMSE: {nmf_results['test_rmse'].mean():.3f}")
print(f"NMF - All RMSE scores: {nmf_results['test_rmse']}")
```

### Summary of Best Scores

```python
print("\n--- Model Performance Summary ---")
print(f"KNNBasic Mean RMSE: {knn_basic_results['test_rmse'].mean():.3f}")
print(f"SVD Mean RMSE: {svd_results['test_rmse'].mean():.3f}")
print(f"NMF Mean RMSE: {nmf_results['test_rmse'].mean():.3f}")

# Note: Surprise's cross_validate does not directly provide 'best params' because it runs with fixed parameters per model initialization.
# To find best params, a GridSearchCV would be needed.
print("\nFor KNNBasic, SVD, and NMF, the parameters used were the ones provided in their initialization.")
print("KNNBasic Parameters: k=20, sim_options={'name': 'msd', 'user_based': True}")
print("SVD Parameters: default (except verbose=False)")
print("NMF Parameters: default (except verbose=False)")
```
